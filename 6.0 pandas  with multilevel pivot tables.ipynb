```python
area = pd.Series({'California': 423967, 'Texas': 695662,
                  'New York': 141297, 'Florida': 170312,
                  'Illinois': 149995})
pop = pd.Series({'California': 38332521, 'Texas': 26448193,
                 'New York': 19651127, 'Florida': 19552860,
                 'Illinois': 12882135})
data = pd.DataFrame({'area':area, 'pop':pop})
data
```

```python
data['density']=data['area']/data['pop']
data
```

```python
x=data.loc[data.density <10000, ['pop', 'density']]
x
```

```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
titanic=sns.load_dataset('titanic')
titanic.head(2)
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic.groupby(['sex','class'])['survived'].aggregate('mean').unstack()
```

```python
x=titanic.groupby(['sex','class'])
print("",*x)
```

```python
titanic.groupby('survived')[['age']].mean()
titanic.groupby('survived')[['age']].max()
```

```python
titanic.groupby('sex')['who'].head()
```

```python
titanic.pivot_table('survived',index='sex',columns='class')
```

```python
titanic.groupby(['sex'])[['age','class']].mean()
titanic.groupby(['sex','class'])['age'].aggregate('mean').unstack()
```

# multilivel pivot table

```python
age = pd.cut(titanic['age'], [0, 18, 80])
```

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```
