# stripplot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=.10,linewidth=0.8)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=.10,linewidth=0.8,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,split=True,\
              palette='winter_r',jitter=1,linewidth=1,hue='sex')
```

```python
sns.stripplot(x='sex',y='tip',hue='smoker',data=tips,\
              jitter=False,marker='d',size=10)
```

```python
sns.stripplot(x='sex',y='tip',hue='smoker',data=tips,\
              jitter=True,marker='d',size=5,alpha=1)
sns.boxplot(x='sex',y='tip',data=tips)
plt.show()
```

# violin plot

```python
sns.stripplot(x='tip',y='day',data=tips,jitter=True)
sns.violinplot(x='tip',y='day',data=tips,color='0.9')
```

# pairgrid

```python
x=sns.PairGrid(iris,hue='species')
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue="species",palette='coolwarm')
x=x.map_diag(plt.hist,edgecolor='red')
x=x.map_offdiag(plt.scatter,edgecolor='green')
```

```python
iris.columns
```

```python
x=sns.PairGrid(iris,hue='species',vars=['petal_length','petal_width'])
x=x.map_diag(plt.hist,edgecolor='red')
x=x.map_offdiag(plt.scatter,edgecolor='black')
x=x.add_legend()
```
